```python
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
tips=pd.read_csv('FIFA 2018 Statistics.csv')
```

```python
tips
```

```python
tips.head()
```

```python
sns.distplot(tips['Attempts'],kde=True)
#A distplot that shows a histogram with a kde curve
```

```python
sns.jointplot(x='Ball Possession %',y='Goal Scored',data=tips)
#compared the distribution for Ball possession vs Goal scored
```

```python
sns.jointplot(x='Ball Possession %',y='Goal Scored',data=tips,kind='reg')
#jointplot to compare Ball Possession attribute with the Goal Scored
```

```python
sns.jointplot(x='Ball Possession %',y='Goal Scored',data=tips,kind='kde')
#goal density using kernel density estimation(kde)
```

```python
sns.rugplot(tips['Attempts'])
#Rugplot for the no. of attempts
```

```python
sns.kdeplot(tips['Attempts'])
#Kernel Density Estimation Plot
```

```python
sns.barplot(x='Attempts',y='Goal Scored',data=tips)
#Average Goal Scored per (no.of Attempts) value
```

```python
import numpy as np
```

```python
sns.barplot(x='Attempts',y='Goal Scored',data=tips,estimator=np.std)
#standard deviation of Goal Scored per (no.of Attempts) value by its mean value
#Barplot
```

```python
import matplotlib.pyplot as plt
```

```python
sns.boxplot(x='Round',y='Goal Scored',data=tips)
plt.tight_layout()
#Box_plot that shows the avg.Goal Scored in every round
```

```python
sns.boxplot(x='Round',y='Goal Scored',data=tips,hue='Man of the Match')
plt.tight_layout()
```

```python
#Violin plots
sns.violinplot(x='Round',y='Goal Scored',data=tips)##
```

```python
sns.violinplot(x='Round',y='Goal Scored',data=tips,hue='Man of the Match',split=True)
```

```python
sns.stripplot(x='Round',y='Goal Scored',data=tips,hue='Man of the Match',split=True)
#STRIPPLOT 
```

```python
sns.violinplot(x='Round',y='Goal Scored',data=tips)
sns.swarmplot(x='Round',y='Goal Scored',data=tips,color='Black')
#To show a better information violinplot is combined with a swarmplot
```

```python
sns.factorplot(x='Round',y='Goal Scored',data=tips,kind='bar')
plt.tight_layout()
plt.tight_layout()
```

```python
sns.lmplot(x='Ball Possession %',y='Goal Scored',data=tips,col='Man of the Match')
```
